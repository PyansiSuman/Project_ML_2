```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('50_Startups.csv',usecols=['R&D Spend','Administration','Marketing Spend','Profit'])
df
```

```python
df.shape
```

```python
df.info()
```

```python
#EDA
col = ['R&D Spend','Administration','Marketing Spend']
for i in col:
  sns.scatterplot(x=df[i],y=df['Profit'])
  plt.show()
```

```python
# input and output column split
x = df.iloc[:,0:3]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# Train - Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#model building
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
x_test.head(2)
```

```python
y_test.head(2)
```

```python
y_pred = model.predict(x_test)
y_pred[:5]
```

```python
profit = model.predict([[65605.48,153032.06,107138.38]])
profit
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
MAE = mean_absolute_error(y_test,y_pred)
MSE = mean_squared_error(y_test,y_pred)
RMSE = root_mean_squared_error(y_test,y_pred)
R2 = r2_score(y_test,y_pred)

print('MAE value:',MAE)
print('MSE value:',MSE)
print('RMSE value:',RMSE)
print('R2 value:',R2)
```
